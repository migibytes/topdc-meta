```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def Qtot(alpha,Q0,Qnr):
    alpha_opt=np.sqrt(Q0/Qnr)
    return Qnr/((alpha**2 / alpha_opt**2) + 1) 

def Qtot2(alpha,Qnr):
    # alpha_opt=np.sqrt(Q0/Qnr)
    alpha_opt = 0.4055
    return Qnr/((alpha**2 / alpha_opt**2) + 1) 

def THG_enhancement(alpha,Q0,Qnr):
    alpha_opt=np.sqrt(Q0/Qnr)
    alpha_p = alpha**2 / alpha_opt**2
    return Qnr**3 * (alpha_p/(alpha_p+1)**2)**3
```

```python
import matplotlib
matplotlib.rcParams['mathtext.fontset'] = 'stix'
from matplotlib.ticker import AutoMinorLocator

asy_range = np.linspace(0.02,0.45)
Q0=8
Qnr=200

fig, ax = plt.subplots(layout='constrained')
sns.set(font_scale=1.25,style='white', palette='deep',font='Garamond',rc = {'figure.figsize':(7,5)})

plt.semilogy(asy_range,Qtot(asy_range,Q0,Qnr),label='$Q_{tot}$',color='k')
plt.semilogy(asy_range,(Q0*asy_range**-2),'--',label='$Q_{rad}$',color='C0')
plt.axhline(Qnr,label=r'$Q_{nr}$',linestyle='-.',color='C0')
plt.axvline(np.sqrt(Q0/Qnr),linestyle=':',alpha=0.8,color='C2',label=r'$\alpha_{opt}$')
plt.plot(np.sqrt(Q0/Qnr),Qtot(np.sqrt(Q0/Qnr),Q0,Qnr),marker='o',markersize=8,color='k',fillstyle='none')
#plt.plot(results['Asymmetry'],THG_conv_eff,marker='o',fillstyle='none',label='THG Conversion')


plt.legend(loc='best',reverse=False)
plt.xlabel(r'Asymmetry ($\alpha$)')
plt.ylabel(r'Q-Factor')
plt.xlim(0.05,0.45)
plt.ylim(2e1,1e4)

def asytoS2(asy):
    return 230-(asy*230)

def S2toasy(S2):
    return (230-S2)/230

secax = ax.secondary_xaxis('top', functions=(asytoS2, S2toasy))
secax.xaxis.set_minor_locator(AutoMinorLocator())
secax.set_xlabel(r'S2 (nm)')

ax2 = ax.twinx()
ax2.plot(asy_range,THG_enhancement(asy_range,Q0,Qnr)/max(THG_enhancement(asy_range,Q0,Qnr)),color='C1',linewidth=1.5)
#ax2.plot(np.sqrt(Q0/Qnr),THG_enhancement(np.sqrt(Q0/Qnr),Q0,Qnr),marker='o',markersize=8,color='k',fillstyle='none')
#ax2.set_ylim([0.5e7,0.3e9])
ax2.set_ylabel('THG Enhancement')

#plt.savefig('Q-Factor_Optimization',dpi=200,bbox_inches='tight')
plt.show()
```
